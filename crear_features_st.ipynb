```python
import pandas as pd
import numpy as np
from datetime import datetime, date
```

```python
def unir_archivos_st(auctions,events,target):

    auctions.columns = ['ref_hash','s_source_id','s_ref_type1','s_t_min','s_t_prom',
                    's_t_max','s_ref_type_7','s_#subastas']
    target.columns = ['ref_hash','segundos']
    
    todo = pd.merge(auctions,events,on='ref_hash',how='outer')
    todo = pd.merge(todo,target,on='ref_hash',how='outer')
    
    t_max = 259200
    vacios = {'s_t_min':t_max, 's_t_prom':t_max, 's_t_max':t_max,
              'e_t_prom': t_max, 'segundos':t_max}
    todo = todo.fillna(vacios).fillna(0)
    return todo
```

```python
auctions = pd.read_csv('auctions_train_1820.csv')
events = pd.read_csv('events_train_1820.csv')
target = pd.read_csv('auctions_test_1820.csv')
todo = unir_archivos_st(auctions,events,target)
todo.to_csv('entrenar_st_1820.csv', index=False)
todo.head()
```

```python
def unir_archivos_sc(events,clicks,installs,target):

    todo = pd.merge(events,clicks,on='ref_hash',how='outer')
    todo = pd.merge(todo,installs,on='ref_hash',how='outer')
    todo = pd.merge(todo,target,on='ref_hash',how='outer')
    
    t_max = 259200
    vacios = {'e_t_prom': t_max, 'c_t_prom':t_max, 'c_timeToClick':t_max,
              'i_t_min':t_max, 'i_t_prom':t_max, 'i_t_max':t_max, 'segundos':t_max}
    todo = todo.fillna(vacios).fillna(0)
    del todo['i_attributed']
    return todo
```

```python
events = pd.read_csv('events_train_1820.csv')
clicks = pd.read_csv('clicks_train_1820.csv')
installs = pd.read_csv('installs_train_1820.csv')
target = pd.read_csv('installs_test_1820.csv')
todo = unir_archivos_sc(events,clicks,installs,target)
todo.to_csv('entrenar_sc_1820.csv', index=False)
todo.head()
```
