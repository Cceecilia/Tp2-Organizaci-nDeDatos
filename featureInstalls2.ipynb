```python
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta
```

```python
installs = pd.read_csv('installs.csv.gzip', compression='gzip')
```

```python
installs = installs.drop(['trans_id','click_hash'], axis=1)
```

```python
installs['created'] = pd.to_datetime(installs['created'])
installs['hora'] = installs['created'].dt.hour
installs['fecha'] = installs['created'].dt.date
```

```python
# installs.groupby('ref_hash').agg({'ref_hash':'count'})
```

```python
installs.head()
```

```python
installs['hora'] = installs['hora'].astype('category')
installs.head()
```

```python
from __future__ import division

aplicaciones = installs.groupby(['hora']).agg({'hora':'count'})
aplicaciones.columns = ['cant']
aplicaciones = aplicaciones.reset_index('hora')
app = {}
ap1 = aplicaciones['hora'].to_list()
ap2 = aplicaciones['cant'].to_list()
tamanio = len(installs)
for i in range(len(ap1)):
    app[ap1[i]] = round(ap2[i]/tamanio,5)
print(app)
```

```python
aplic = installs['hora'].value_counts(normalize=True).nlargest(10)
aplic
```

```python
atribuidos = {'attributed': {True:'1',False:'0'}}
nuevo_installs = installs.replace(atribuidos)
nuevo_installs['hora'] -= nuevo_installs['hora'].replace(aplic)
nuevo_installs.head()
```
